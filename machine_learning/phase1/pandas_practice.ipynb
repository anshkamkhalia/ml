```python
import numpy as np

age = df["age"]

condlist = [
    age < 13,
    (age >= 13) & (age < 20),
    (age >= 20) & (age < 60),
    age >= 60
]

choices = ["Child", "Teen", "Adult", "Senior"]

df["age_group"] = np.select(condlist, choices, default="Unknown")

print(df[["age", "age_group"]].head(10))
```

```python
def find_family_size(row):
    return row["sibsp"] + row["parch"] + 1

df["family_size"] = df.apply(find_family_size, axis=1)

df["family_size"].head(10)
```
