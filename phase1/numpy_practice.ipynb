```python
# üî¢ NumPy Practice Problem 2 (Easy-Medium):
# Create a 3x3 NumPy array with values from 1 to 9 (inclusive). Then, calculate:

# The sum of all elements.

# The sum of each row.

# The sum of each column.
```

```python
import numpy as np

array = np.arange(1, 10).reshape(3, 3)
print(array)

total = np.sum(array)
print("Total sum:", total)

row_sums = np.sum(array, axis=1)
print("Row sums:", row_sums)

col_sums = np.sum(array, axis=0)
print("Column sums:", col_sums)
```

```python
# Problem 4 (Intermediate):
# Create a 4x4 NumPy array filled with random integers from 10 to 50 (inclusive).

# Then:

# Replace all values greater than 30 with 30.

# Print the original array and the modified array.
```

```python
arr = np.random.randint(10, 51, size=(4, 4))

for x in range(arr.shape[0]):  
    for y in range(arr.shape[1]):  
        if arr[x][y] > 30:
            arr[x][y] = 30
    
original = np.random.randint(10, 51, size=(4, 4))

print("Original:\n", original)
print("Modified:\n", arr)
```

```python
# Problem 6 (Harder):
# Create a 6√ó6 NumPy array of random integers from 1 to 100.

# Then:

# Replace all even numbers with 0

# Replace all odd numbers with 1

# Now count how many 1s and 0s are in the array

# Print the modified array and the counts

# This one's all about indexing, slicing, and counting like a NumPy wizard. üßô‚Äç‚ôÇÔ∏è

# Let me know when you‚Äôre done, and we‚Äôll tackle #7 next!
```

```python
array = np.random.randint(1,101,size=(6,6))

array[array%2==0] = 0
array[array%2!=0] = 1

odd_count = 0
even_count = 0

for x in range(array.shape[0]):
    for y in range(array.shape[1]):
        if array[x][y] == 0:
            even_count+=1
        else:
            odd_count+=1

print(f"Modified Array:\n {array}")
print(f"Even count {even_count}")
print(f"Odd count {odd_count}")
```

```python
# Problem 7 (Hard):
# Create a 6√ó6 NumPy array filled with random integers from 1 to 100.

# Then:

# Find the maximum value in each row

# Find the index (column) of the max value for each row

# Replace each max value in the array with -1

# Print the modified array
```

```python
import numpy as np

array = np.random.randint(1, 101, size=(6, 6))
print("Original Array:\n", array)

max_vals = np.max(array, axis=1) 

max_indices = np.argmax(array, axis=1)

for i in range(array.shape[0]):
    array[i, max_indices[i]] = -1

print("\nModified Array:\n", array)
```

```python
# üß† Problem 8 (Boss Level):
# You‚Äôre going to build a mini game board.

# Create a 10√ó10 NumPy array filled with 0s

# Randomly place 20 ones (1s) in random positions

# Then, replace every 0 that touches a 1 (up/down/left/right) with a 2 (but not diagonally!)

# Print the final board
```

```python
import numpy as np

board = np.zeros((10, 10), dtype=int)

ones_placed = 0
while ones_placed < 20:
    x = np.random.randint(0, 10)
    y = np.random.randint(0, 10)
    if board[x, y] == 0:
        board[x, y] = 1
        ones_placed += 1
        
positions = np.argwhere(board == 1)

for (x, y) in positions:
    if x > 0 and board[x-1, y] == 0:
        board[x-1, y] = 2
    if x < board.shape[0] - 1 and board[x+1, y] == 0:
            board[x+1, y] = 2
    if y > 0 and board[x, y-1] == 0:
        board[x, y-1] = 2
    if y < board.shape[1] - 1 and board[x, y+1] == 0:
        board[x, y+1] = 2


print("Final Game Board:\n", board)
```
