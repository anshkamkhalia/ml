```python
from sklearn.ensemble import RandomForestClassifier
```

```python
def try_params(**params):
    model = RandomForestClassifier(**params, n_jobs=-1, random_state=42)
    model.fit(X_train, X_train_target)
    train_preds = model.predict(X_train)
    val_preds = model.predict(X_val)
    return {"Params": [params], 
            "Train acc": accuracy_score(X_train_target, train_preds), 
            "Val acc": accuracy_score(X_val_target, val_preds),
            "Estimators": model.n_estimators,
            "Depth": model.max_depth,
            "Model": model}

```

```python
from xgboost import XGBClassifier
```

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
X_train_target = le.fit_transform(X_train_target)
X_val_target = le.fit_transform(X_val_target)  
```

```python
def try_params(**params):
    model = XGBClassifier(**params, n_jobs=-1, random_state=42)
    model.fit(X_train, X_train_target)
    train_preds = model.predict(X_train)
    val_preds = model.predict(X_val)
    return {"Params": [params], 
            "Train acc": accuracy_score(X_train_target, train_preds), 
            "Val acc": accuracy_score(X_val_target, val_preds),
            "Estimators": model.n_estimators,
            "Depth": model.max_depth,
            "Model": model}
```

```python
result = try_params(learning_rate=0.1, n_estimators=350, max_depth=7)
result
```

```python
model = result["Model"]
```

```python
final_preds = model.predict(X_test)
preds = le.inverse_transform(final_preds)

submission = pd.DataFrame({
    "id": X_test_id,        # same order as in test.csv
    "Personality": preds        # your model's predictions
})

submission.to_csv("submission.csv", index=False)
```
